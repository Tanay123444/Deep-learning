## Binary classification - Booking Cancellation 

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

## Read the train dataset

```python
import pandas as pd
df = pd.read_csv('train_booking.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

## Seperate X and Y 

```python
X = df.drop(columns=['id', 'booking_status'])
Y = df[['booking_status']]
```

```python
X.head()
```

```python
Y.head()
```

## Create a preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='median')),
                           ('scaler', StandardScaler())])
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre
```

```python
cols = num_pipe.get_feature_names_out()
cols
```

```python
X_pre = pd.DataFrame(X_pre, columns=cols)
X_pre.head()
```

## Apply Train Test Split 


```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

## Create a Neural Network 

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
model = Sequential()

# Hidden Layer 1 with 17 input neurons
model.add(Dense(64, activation='relu', input_dim=xtrain.shape[1]))
model.add(Dropout(0.4))

# Hidden Layer 2
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.3))

# Output Layer Binary Classification - Sigmoid
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

## Compile the Model 

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Fit your model

```python
hist = model.fit(xtrain, ytrain, validation_split=0.2, epochs=100)
```

## Plot the learning curve for the above model 

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'], label='Training Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Cross Entropy Loss')
plt.title('Learing Curve')
plt.legend()
plt.show()
```

## Evaluate the model 

```python
model.evaluate(xtrain, ytrain)
```

```python
model.evaluate(xtest, ytest)
```

## Predict the probabilities 

```python
yprob_tr = model.predict(xtrain)
yprob_ts = model.predict(xtest)
```

## Get the prediction 

```python
ypred_tr = [1 if i>=0.5 else 0 for i in yprob_tr]
ypred_tr[0:5]
```

```python
ytrain.head()
```

```python
ypred_ts = [1 if i>=0.5 else 0 for i in yprob_ts]
ypred_ts[0:5]
```

```python
ytest.head()
```

## Create the confusion matrix 

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_ts)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

## Check the classification report 

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_ts))
```
