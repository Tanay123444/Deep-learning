## Image Processing 

```python
import cv2 
img = cv2.imread('CuteCat.jpeg')
img
```

```python
img.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

## Cv2 Considers image as BGR 

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img2
```

```python
plt.imshow(img2)
```

## Split above image in rgb components 

```python
r,g,b = cv2.split(img2)
```

```python
r
```

```python
g
```

```python
b
```

```python
plt.imshow(r, cmap = 'Reds')
```

```python
plt.imshow(g, cmap='Blues')
```

```python
plt.imshow(b, cmap= 'Greens')
```

## Reading image as Grayscale 

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

## Reading a GrayScale Image Directly

```python
A = cv2.imread("Alphabet Example.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
A
```

```python
plt.imshow(A,cmap = 'gray')
```

```python
import seaborn as sns 
plt.figure(figsize=(15,15))
sns.heatmap(A, annot = True, fmt = 'd')
plt.show()
```
